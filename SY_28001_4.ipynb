```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io

from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['StudentsPerformance (4).csv']))
print(df)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
mean = df.mean(numeric_only=True)
print("MEAN : ")
print(mean)
```

```python
median = df.median(numeric_only=True)
print("MEDIAN : ")
print(median)
```

```python
mode = df.mode()
print("MODE : ")
print(mode)
```

```python
variance = df.var(numeric_only=True)
print("Variance : ")
print(variance)
```

```python
StandardDev = df.std(numeric_only=True)
print("Standard Deviation : ")
print(StandardDev)
```

```python
quantiles = df['math score'].quantile([0.25, 0.50, 0.75])

print(quantiles)
```

```python
print(df.dtypes)
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=df)
plt.xlabel('math score')
plt.ylabel('writing score')
plt.title('Boxplot : math score VS writing score')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(numeric_only=True),annot=True, cmap='coolwarm')
plt.title('Correlation Matrix : ')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['writing score'], df['math score'], alpha=0.5)

plt.title("Math Score vs Writing Score", fontsize=16)
plt.xlabel("Writing Score", fontsize=12)
plt.ylabel("Math Score", fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.histplot(df['math score'], kde=True, color='pink', bins=10)
plt.title("Normal Distribution - MATH Scores")
plt.xlabel("Score")
plt.xlabel("Frequency")
plt.show()
```
